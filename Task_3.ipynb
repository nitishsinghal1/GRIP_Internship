### Exploratory Data Analysis

```python
# heatmap to find correlation between attributes
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
```

```python
# histogram
df.hist(figsize=(10,10),color="red")
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Segment',data=df, palette='rainbow')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Category',data=df, palette='rainbow')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Sub-Category',data=df, palette='rainbow')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Ship Mode',data=df, palette='rainbow')
```

```python
# distplot
sns.distplot(df['Profit']);
```

```python
# scatterplot
sns.scatterplot(df['Profit'], df['Sales'], hue = df['Ship Mode'])
```
